```python
import numpy as np
from scipy.io import wavfile
l=25
n=10000
fs=10000
mean=0
sd=1
volume=2000

def white_noise_generator(mean,sd):
    noise=np.random.normal(mean,sd,size=l)
    np.array(noise)
    return noise

def filter_lp(signal):
    newsig=np.roll(signal,1)
    arr=(newsig+signal)/2
    return arr

#main
a=white_noise_generator(mean,sd)
buff = a

while len(a)<n:
     buff = filter_lp(buff)
     a=np.append(a,buff,axis=0)   

a*=volume

wavfile.write("testing.wav",fs,np.int16(a))
```
